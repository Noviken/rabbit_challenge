### ロジスティック回帰　レポート

- 分類問題
    - 回帰問題では、連続値を予測していたが、分類問題では離散値を予測
        - 2値分類であれば、生きているのか死んでいるのか、花なのかそうではないのか…とか。
    
- ロジスティック線形回帰モデル
    - 分類問題を解くための教師あり学習モデル
        - 入力データとパラメタベクトルとの線形結合をシグモイド関数の入力値とし、その値をy=1である確率として扱う
    - シグモイド関数
        - 必ずその値は0から1の間
        - 単調増加関数
        - シグモイド関数の微分をシグモイド関数で表すことが可能のため、尤度関数の微分の際の計算が容易

        $$
        \begin{align}
            \sigma(x) &= \frac{1}{1+\exp(-ax)} \\
            \frac{\partial \sigma(x)}{\partial x}&= a \sigma(x) (1 - \sigma(x))　\\
        \end{align} 
        $$

    - ロジスティック回帰モデルでは、確率分布としてベルヌーイ分布を利用
        - ベルヌーイ分布：数学において、確率pで1をとり、確率1-pで0をとる離散確率分布。
    - ベルヌーイ分布のパラメタの推定
        - ある分布を考えたとき、そのパラメタによって生成されるデータは変化
        - データからそのデータを生成したであろう尤もらしい分布(パラメタ)を推定したい → 最尤推定
    - 尤度関数が最大となるパラメタを探す
        - 対数をとると、微分が簡単(同時確率の積が、和の変換できる.指数部も積の形の変換できる)
        - 対数関数は単調増加関数なので、対数尤度関数が最大になる点と、尤度関数が最大になる点は同じ

- どうやって最適なパラメタを探す
    - 対数尤度関数をパラメタで微分して0になる解を求めたいが、解析的に解くのは困難
    - (バッチ)勾配降下法
        - 反復学習によりパラメタを逐次的に更新するアプローチの１つ
        - $\eta$は学習率と呼ばれるハイパーパラメタで、収束のし易さを調整
            $$
                w_{k+1} = w_k - \eta\frac{\partial E(w)}{\partial w}
            $$
        - パラメタが更新されなくなった＝勾配が0。少なくとも反復学習で探索した範囲では最適な解となった
            - でも実際は鞍点だったりする可能性もある。
        - 勾配降下法は、パラメタ更新するのに全データに対する和を求める必要がある
            - データ数が膨大な場合、メモリが不足。計算時間も膨大。
            - 確率的勾配降下法（やミニバッチ勾配降下法）で解決
    - 確率的勾配降下法
        - ランダムで（確率的に）1つのデータを選び、そのデータをもとにパラメタを更新
        - 勾配降下法でパラメタ更新1回する計算量でn回更新できる
        
- 分類問題の性能指標
    - 正解率
        - 読んで字の如く、正解した割合。母数は推測した全データ数
    - 再現率(recall)
        - 本当にPositiveなものを、どれだけPositiveと推測できたかの割合
        - 例え偽陽性が高くとも、取りこぼしがないようにしたい場合に利用
    - 適合率(precision)
        - Positiveと推測したもののうち、本当にPositiveである割合
        - 取りこぼしが多くとも、より精確な予測をしたい場合に利用。スパムメール検出とか。
    - f値
        - 再現率と適合率の調和平均であり、両者のバランスを示す。高ければ高いほど、両者も高い。
        - 再現率と適合率はトレードオフの関係にある
