### 非線形回帰　レポート

- 複雑な非線形構造を似剤する現象に対して、非線形回帰モデリングを実施
    - 非線形…線形ではない。すなわち加法性、斉次性を満たさない
    - データの構造を線形で捉えられる場合は限られる　→　非線形な構造を捉えられる仕組みが必要
        - 例えば…降伏点以降の応力ひずみ関係とか。数学的だけならば二次関数でも指数関数でも三角関数でも。

- 基底展開法
    - 回帰関数として、基底関数と呼ばれる既知の非線形関数とパラメタベクトルの線形結合を使用
    - 未知のパラメタは、最小２乗法や最尤法により推定（線形回帰と同じ）
    - よく使われる基底関数
        - 多項式関数
        - ガウス型基底関数
        - スプライン関数 / Bスプライン関数
    - 基底関数法も、線形回帰と同じ枠組みで推定が可能
    
- 未学習と過学習
    - 未学習：学習データに対して、十分小さな誤差が得られないモデル。またはその状態。
        - この場合、学習データ数の不足やエポックの不足ではなく、表現力が不足。より表現力の高いモデル活用の検討要
    - 過学習：学習誤差は小さくなったが、検証誤差が大きいモデル
        - 対策としては、学習データの数を増やす、モデルの表現力を低くする
        - 表現力を低くするには：不要な基底関数を削除、正則化を利用（e.g. ラッソ、リッジ、…）
    - 適切な表現力のモデルが望ましい
        - 適切な表現力のモデル＝汎化性能が高いモデル
            - 汎化性能が高いことはどうやって評価する？→交差検証
- 汎化性能
    - 本来高くあって欲しい、未知の入力データに対する予測性能
    - 通常、学習データとは別に収集された検証データでの性能を測ることで推定する
    
- 評価方法
    - ホールドアウト法
        - 有限のデータを学習用とテスト用の２つに分割して使用
            - 学習用に多く用いれば性能評価の精度は悪くなり、その逆もまた然り。
            - 上記のように、手元に大量のデータがある場合を除き、良い性能評価を与えない欠点がある
        - 基底展開法に基づく非線形回帰モデルでは、基底関数の数や位置、バンド幅の値とパラメタを、ホールドアウト値を小さくするモデルで決定
        
    - 交差検証（クロスバリデーション）
        - データをk個のグループに分割
        - 1グループを検証用に用いることとし、残りでモデルを学習。精度を評価。
        - 上記をグループの全パターン（各グループが1回ずつ検証用となる）実施
        - 各試行における精度の平均値を以ってモデルの評価値とする
