### SVM　レポート

- #### SVM
    - サポートベクターマシンは応用範囲が広く、分類・回帰どちらにも用いることができる
        - このレポートでは、２値分類に利用する場合について記載
        
- #### 線形サポートベクターマシン
    - 線形サポートベクターマシンは、マージン最大化という方針で、データ群からできるだけ離れた決定境界を学習する
    - ロジスティック回帰同様、線形な決定境界となるが、ロジスティック回帰より良い境界を得られる場合がある
        - マージン最大化という考え方により、決定境界ができるだけデータから離れるように求まるため
        - マージンとは、決定境界と、決定境界に最も近いデータとの間の距離
    - もしデータ（２クラス）が、線形分離できなければ？
        - (線形SVCの範囲では)ソフトマージンを導入する
            - ハードマージン：マージンの内側にデータが入り込むことを許容しない
            - ソフトマージン：マージンの内側にデータが入り込むことを許容する
                - どの程度許容するのかはハイパーパラメタであり、決定には交差検証などで検討が必要
    - 線形分離可能であれば、ハードマージンが絶対的に良いのか？
        - No.データに対して過学習してしまう場合（外れ値に引きずられ、自然ではない決定境界となる可能性など）がある
        
- #### カーネル法
    - カーネル法を利用することによって、サポートベクターマシンで複雑な（線形ではない）決定境界を学習することができる
    - 学習データを高次元空間からの射影の結果であり、その高次元空間では線形分離可能だと考える
    - そのような空間でサポートベクターマシンを用いれば決定境界を学習できるはず
    - どうやって、そのような高次元空間を構成するのか？
        - カーネル法では、カーネル関数と呼ばれる関数を利用することで、空間を具体的に構築することなく決定境界を利用できる
    - カーネル関数
        - カーネル関数には様々なものがあるが、利用するカーネル関数によって得られる決定境界が変わる
        - 代表的なカーネル関数
            - 線形カーネル（線形サポートベクターマシンと同じ）
            - シグモイドカーネル
            - 多項カーネル
            - RBFカーネル
