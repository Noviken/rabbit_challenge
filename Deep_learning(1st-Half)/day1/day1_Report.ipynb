### 深層学習（day1)　レポート

- #### Section1 入力層〜中間層
    - 入力層は、NNに何らかの数字（データ）を入れる部分.また、それぞれのデータを受け取る部分をノードと呼ぶ。
    - それぞれのデータをどれほど使うか、を示すパラメタを重み（weight)と呼ぶ。
    - 重みでは対応できない、データ全体に何かを足し合わせたりするような効果をもつパラメタをバイアス(bias)と呼ぶ。
    - 中間層は、W,bによって変換されたデータを活性化関数の入力として演算し、次の層へ出力する層
        
- #### Section2 活性化関数
    - ニューラルネットワークにおいて、次の層への出力の大きさを決める”非線形の”関数
    - 入力値の値によって、次の層への信号のON/OFFや強弱を定め働きを持つ
    - 非線形であることが肝要。線形をいくら重ねたところで、単純パーセプトロンと同義
    - 中間層用の活性化関数
        - ReLU : max(0,x)。　勾配消失問題の回避とスパース化に貢献することで良い成果をもたらしている
        - シグモイド関数 : 0-1の間を緩やかに変化する関数。大きな値では勾配が微小であるため、勾配消失問題を引き起こすことがあった
        - ステップ関数 : x>0 ならば１、x<=0ならば0を出力する関数.0-1の間を表現できず、線形分離可能なものしか学習できなかった
    - 出力層用の活性化関数
        - ソフトマックス関数　：　足して1となる結果を出力。多クラス分類などで用いられる。
        - 恒等写像 : 入力をそのまま出力。回帰問題で用いられる。
        - シグモイド関数 : 関数そのものについて記載は割愛。２クラス分類などで用いられる。

- #### Section3 出力層
    - 誤差関数
        - 出力層からの出力が、正解（教師データ）と比較してどの程度誤っているかの尺度を与える関数
        - 誤差関数で評価した誤差が小さくなるようにモデルを学習させていく
        - 回帰ではMSE、分類ではクロスエントロピーなどが用いられる
    - 出力層の活性化関数
        - 中間層との違い
            - 値の強弱　：　出力層では信号の大きさ（比率）はそのままに変換
            - 確率出力:分類問題の場合、出力層の出力は0-1の範囲に限定し、総和を1とする必要がある
    
- #### Section4 勾配降下法
    - 勾配降下法
        - パラメタ最適化手法の1つ
        - 誤差関数のパラメタに対する微分（勾配）に学習率をかけたものを減算することにより、パラメタを更新する
        - 全データ文の誤差を用い、１度にパラメタを更新する
    - 確率的勾配降下法（SGD）
        - 勾配降下法のパラメタ更新単位を、全データではなくrandomに選択した1データにしたもの
        - 1データによる更新となるため、オンライン学習が可能となる。またメモリスペースも少なく済む。
        - また、randomnessが加わるため、単なる勾配降下法と比べて局所解にトラップされない可能性がある。
    - ミニバッチ勾配降下法
        - 全データを１度に用いるのではなく、ミニバッチと呼ぶ部分集合に分割して学習・パラメタ更新に用いる
        - ミニバッチサイズが1の場合、それはSGDと同じ
    
- #### Section５ 誤差逆伝搬法
    - 直接微分を計算すると、計算負荷が大きく難あり。そこで、誤差逆伝搬法を利用する
    - 誤差逆伝搬法：算出された誤差を、出力層側から順に微分して前の層へと伝搬させていき、最小限の計算で各パラメタでの微分値を”解析的に”計算する手法
    - 不要な再帰的計算を避けて微分を算出することができる
